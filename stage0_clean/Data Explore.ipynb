```python
# Number of earthquake articles

for article in cps_withtags.find( {'headline': {'$regex': '.*earthquake.*'}}, {'headline':1, 'linkData':1} ):
    pprint.pprint(article)
```

```python
cps_withtags.count_documents({'headline': {'$regex': '.*earthquake.*'}})
```

```python
cps_withtags.count_documents({'headline': {'$regex': '.*tsunami.*'}})
```

```python
# Number of tsunami articles

for article in cps_withtags.find({'headline': {'$regex': '.*tsunami.*'}}, 
                                {'headline':1, 'firstPublished':1}):
    pprint.pprint(article)
```

```python
tot_count = cps_withtags.count_documents({})
```

```python
earthquake_count = cps_withtags.count_documents({'headline': {'$regex': '.*earthquake.*'}})
```

```python
earthquake_count/tot_count * 100
```

```python
# How many earthquakes are breaking news?

breaking_earthquake_count = cps_withtags.count_documents( {'$and': [{'headline': {'$regex': '.*earthquake.*'}},
                                                                   {'options.isBreakingNews': True}]})
    
```
