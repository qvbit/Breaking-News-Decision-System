```python
import pandas as pd
import numpy as np
import os
from sutime import SUTime
import datetime
```

```python
df = pd.read_pickle('../dataframes/df_eq_label.pkl')
```

```python
jar_files = os.path.join(os.path.dirname('/Users/naumanw1/Documents/fod/jars'), 'jars')
sutime = SUTime(jars=jar_files, mark_time_ranges=True)
```

```python
def eq_only(df):
    df = df[df['label'] == 1]
    df.dropna(subset=['body'], inplace=True)
    
    return df.reset_index()

df = eq_only(df)
```

```python
df = df[['body', 'firstPublished']]
docs = list(zip(df['body'], df['firstPublished']))
```

```python
dates = []
date_words = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']

for doc, ref in docs:
    doc = doc.replace('(', '').replace(')', '')
    ref = str(ref).split(' ')[0] + 'T' + str(ref).split(' ')[1][:5] + '+0000'
    
    times = sutime.parse(doc, reference_date=ref)
    
    if times:
        found = False
        for i, time in enumerate(times):
            if 'GMT' in time['text'] or 'BST' in time['text']:
                dates.append(time['value'])
                found = True
                break
        if not found:
            for i, time in enumerate(times):
                if any(date_word in time['text'].lower() for date_word in date_words):
                    dates.append(time['value'])
                    found = True
                    break
        if not found:
            dates.append( str(ref).split(' ')[0] + '(not found)')
    else:
        dates.append( str(ref).split(' ')[0] + '(not found)')
```
